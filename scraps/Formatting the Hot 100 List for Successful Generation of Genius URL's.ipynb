This notebook gets the first artist from the list of artist names, formats these for URL generation for extraction from Genius.com, and also reformats the title for the same purpose.

The cells for extraction of first artist were created by Khyati Malik and Jonghoon Shin and then modified by Ritwik Banerji.

The cells for reformatting of title track names were created by Ritwik Banerji.

The primary purpose of this notebook is to properly format the data from the Billboard charts so this can be used to successfully generate urls to use to extract lyrics from Genius.com

```python
import pandas as pd
import numpy as np
```

```python
import os
path = os.getcwd()
path
```

```python
filname = '/hot_100_with_lyrics.csv'
hot100 = pd.read_csv(path+filname)
hot100.head()
```

```python
hot100new = hot100.copy()
art_name = hot100new['artists'].tolist()
```

```python
## This cell leaves just the first artist's name.
## It also removes punctuation in accordance with the way that Genius
## URL's are formatted.

first = []
for art in art_name:
    art = art.rsplit(sep='Featuring')[0].strip()
    art = art.rsplit(sep='With')[0].strip()
    art = art.rsplit(sep='&')[0].strip()
    art = art.rsplit(sep='+')[0].strip()
    art = art.replace('$','s')
    art = art.replace('P!nk','Pink')
    art = art.replace('!',' ')
    art = art.replace('*',' ')
    art = art.replace('.','')
    art = art.replace('-',' ')
    art = art.replace('  ',' ')
    art = art.lower()
    first.append(art)
type(first)
```

```python
hot100new['first_artist'] = first
```

```python
hot100new.head()
```

Now we have csv with the artist names formatted. We now need a csv with both the artist and title's formatted to conform to the conventions used by Genius.com URL's.

Run a similar loop to the one above for artist names but this time for track titles.

```python
## make a new dataframe for the formatted tracks and titles
hot100form = hot100.copy()

## make a list called 'hits' consisting of all the track titles.
hits = hot100new['title'].tolist()
hits
```

```python
## This cell reformats track titles so they generate successful Genius.com urls.
## use the list of procedures a few cells above this one to properly format the titles.

## great an empty list of formatted ('fmtd') track names
title_fmtd = []
for title in hits:

    ## First make the title lower case:
    title = title.lower()
    ## remove , ‘ ( ) ? ! . 
    title = title.replace(',','')
    title = title.replace('\'','')
    title = title.replace('(','')
    title = title.replace(')','')
    title = title.replace('?','')
    title = title.replace('!','')
    title = title.replace('.','')
    ## replace ‘ & ‘ with ‘-’
    title = title.replace(' & ', ' and ')
    ## replace ‘&’ with ‘-‘
    title = title.replace('&', '-')
    
    title_fmtd.append(title)
title_fmtd
```

The above is a list of all the track names in the dataset in lowercase with no punctuation.

Now we load this back into our dataframe:

```python
hot100new['title_fmtd'] = title_fmtd
hot100new
```

Please note: the title will still need hyphens inserted in order to successfully generate urls.
