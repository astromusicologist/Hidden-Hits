### Get's and normalizes the artist names from `hot_100_first_artist_v1.csv`

```python
import pandas as pd
import numpy as np
```

```python
import os
path = os.getcwd()
path
```

```python
filname = '\\hot_100_first_artist_v1.csv'
dt = pd.read_csv(path+filname)
dt.head()
```

```python
dtnew = dt.copy()
art_name = dtnew['first_artist'].tolist()
```

```python
first = []
for art in art_name:
    art = art.rsplit(sep='Featuring')[0].strip()
    art = art.rsplit(sep='&')[0].strip()
    art = art.rsplit(sep='+')[0].strip()
    art = art.replace('$','s')
    art = art.replace('P!nk','Pink')
    art = art.replace('!',' ')
    art = art.replace('*',' ')
    art = art.replace('.',' ')
    art = art.replace('-',' ')
    art = art.replace('  ',' ')
    first.append(art)
first
```

```python
dtnew['first_artist_nosp'] = first
```

```python
dtnew.head()
```

```python
dtnew.to_csv(path+'\\hot_100_first_artist_v2.csv')
```
